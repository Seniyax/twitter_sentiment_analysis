```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
df = pd.read_csv('/content/clean_data.csv')
df.head()
```

```python
def clean_dataset(text):
  text = text.lower()

  text = re.sub(r'http\S+|www\S+|https\S+','',text,flags = re.MULTILINE)
  text = re.sub(r'@\W+','',text)
  text = re.sub(r'[^\w\s.,!?]', '', text)
  text = re.sub(r'\s+','',text).strip()

  return text

df['text'] = df['text'].apply(clean_dataset)
```

```python
df.isna().sum()
```

```python

df['label']=np.nan_to_num(df['label'],nan=0)
df.isna().sum()
```

```python

df['label'] = df['label'].astype(int)

```

```python
df.head()
```

```python
df['text']= df['text'].apply(word_tokenize)
```

```python
df['text'] = df['text'].apply(lambda x: ''.join(x))
```

```python
vectorize = CountVectorizer()
X_real = vectorize.fit_transform(df['text'])
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_real,df['label'],test_size=0.2,random_state=42)
```

```python
logic = LogisticRegression(multi_class='multinomial', solver='lbfgs')
logic.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = logic.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(f'Accuracy: {accuracy}')
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = rf.predict(X_test)
accuracy = accuracy_score(y_test,y_pred)
print(f'Accuracy: {accuracy}')
```
